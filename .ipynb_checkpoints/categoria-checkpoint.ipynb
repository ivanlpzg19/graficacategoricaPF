```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_students=pd.read_csv("C:\\Users\\ilopezg\\Downloads\\clean_students_complete.csv")
data_students.head()
```

```python
data_students["cat_read"]=np.where(data_students["reading_score"]>=70, "aprobado", "reprobado")
data_students["cat_math"]=np.where(data_students["math_score"]>=70, "aprobado", "reprobado")
```

```python
data_students
```

```python
aproval=data_students.groupby(["grade","school_name"])["cat_read"].value_counts()
aproval
```

```python
#data_students.to_csv("clean students categorical.csv")
aproval.plot(kind="bar",color='#5F9EA0',figsize=(30,8))
plt.title("Alumnos aprobados y reprobados por escuela y grado")
plt.ylabel("Alumnos")
```
